18.จงเขียนโปรแกรมสุ่มเลขจำนวนเต็มในช่วง 10-100 ออกมาอยู่ใน matrix 5x5 และใช้ If-else เช็คทีละตัวว่าตัวใดบ้างที่มีค่ามากกว่า 50 และปริ้นออกมา

```python
arr18 = np.random.randint(10,101,(5,5))
print(arr18)
rows = len(arr18)
columns = len(arr18[0]) 
for r in range(rows):
    for c in range(columns):
        if arr18[r][c] > 50:
            print(arr18[r][c])
```

20.เขียนโปรแกรมรับค่าจากผู้ใช้ให้ป้อนเลข 9 ตัว แต่ละตัวต้องไม่ซ้ำกันและมีค่าอยู่ในช่วง 10-100 นำเลข 9 ตัวนั้นมาทำเป็น matrix 3x3 และ transport และหาเลขอยู่ตำแหน่งที่ (2,2)

```python
lst = [] 
for i in range(9):
    while 1 :
        num1 = int(input('input number : '))
        if (num1 >= 10 and num1 <= 100 and num1 not in lst):
            lst.append(num1)
            break
        
print('\nMatrix 3x3\n',np.array(lst).reshape(3,3))
print('\nTransport\n', np.array(lst).reshape(3,3).T) 
print('\nArray (2,2)\n', np.array(lst).reshape(3,3).T[2,2]) 
```
